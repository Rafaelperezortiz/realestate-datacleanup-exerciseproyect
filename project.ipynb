#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

casas = pd.read_csv('assets/real_estate.csv', sep=';')

municipios = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
fig, ax = plt.subplots()

for municipio in municipios:
    propiedades_municipio = casas[casas['level5'] == municipio]
    normalized_prices = (propiedades_municipio['price'] - propiedades_municipio['price'].mean()) / propiedades_municipio['price'].std()
    ax.hist(normalized_prices, bins=30, alpha=0.5, label=municipio)

ax.set_xlabel('Normalized Price')
ax.set_ylabel('Frequency')
ax.set_title('Normalized Prices Histogram')
ax.legend()
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
import pandas as pd
casas = pd.read_csv('assets/real_estate.csv', sep=';')
municipios = ['Getafe', 'Alcorcón']
surmadrid = casas[casas['level5'].isin(municipios)]

surmadrid['pps'] = surmadrid['price'] / surmadrid['surface']

getafe_data = surmadrid[surmadrid['level5'] == 'Getafe']
alcorcon_data = surmadrid[surmadrid['level5'] == 'Alcorcón']

mean_pps_getafe = getafe_data['pps'].mean()
mean_pps_alcorcon = alcorcon_data['pps'].mean()

print(f"Precio Promedio por Metro Cuadrado en Getafe: {mean_pps_getafe:.2f} €/m²")
print(f"Precio Promedio por Metro Cuadrado en Alcorcón: {mean_pps_alcorcon:.2f} €/m²")
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Cargar los datos (asumiendo que ya tienes el DataFrame "casas")
casas = pd.read_csv("assets/real_estate.csv", sep=";")

# Definir las cuatro poblaciones (puedes ajustarlas según sea necesario)
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Crear subtramas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(24, 16))

# Iterar a través de las poblaciones y crear gráficos de dispersión
for i, poblacion in enumerate(poblaciones):
    fila, columna = i // 2, i % 2
    subconjunto = casas[casas["level5"] == poblacion]
    axes[fila, columna].scatter(subconjunto["longitude"], subconjunto["latitude"], label=poblacion, alpha=0.5)
    axes[fila, columna].set_title(poblacion)
    axes[fila, columna].set_xlabel("Longitud")
    axes[fila, columna].set_ylabel("Latitud")

# Agregar leyenda
fig.legend(loc='upper right')

# Ajustar el diseño
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
from ipyleaflet import Map, Marker
import numpy as np
import pandas as pd

# Cargar los datos
casas = pd.read_csv("assets/real_estate.csv", sep=";")
municipio = ("Fuenlabrada", "Leganés", "Getafe", "Alcorcón")

# Filtrar las filas que corresponden a los municipios en "municipio"
surmadrid = casas[casas["level5"].isin(municipio)]

# Seleccionar las columnas "latitude" y "longitude" para las coordenadas
coordenadas = surmadrid[["latitude", "longitude"]]

# Agregar marcadores al mapa
for index, row in coordenadas.iterrows():
    latitud, longitud = float(row["latitude"].replace(",", ".")), float(row["longitude"].replace(",", "."))
    marker = Marker(location=(latitud, longitud))
    map.add_layer(marker)

# Mostrar el mapa
map
```
